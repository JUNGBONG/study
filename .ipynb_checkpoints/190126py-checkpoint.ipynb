```python
def solution(arrA, arrB):
    if bool(set(arrA)-set(arrB))==False:
        a= arrA[0]
        count=0
        for i in arrB:
            if i==a:
                if arrA == arrB[count:]+arrB[0:count]:
                    answer=True
                    return answer
                else:
                    count+=1
            else:
                count+=1
    else:
        answer= False
    
    return answer

# arrA=[7, 8, 10]
# arrB=[10, 7, 8]
arrA =[4, 3, 2, 1]
arrB =[5, 4, 1, 2]
solution(arrA, arrB)
```

이때, d 값이 충분히 크다면 전체 도로를 밝게 비출 수 있지만, d 값이 작아진다면 도로 위에 빛이 닿지 않는 부분이 생길 수도 있습니다. 따라서, 도로 위에 어두운 부분이 생기지 않도록 하는 d 값 중 최솟값을 구하려고 합니다. 전체 도로의 길이 l, 가로등이 세워져 있는 위치가 들어있는 배열 v가 매개변수로 주어질 때, 위의 모든 조건을 만족하는 d 의 최솟값을 return 하도록 solution 함수를 완성해주세요.

```python
import math
def solution(l, v):
    n=len(v)
    a=v
    a.append(l)
    a.append(0)
    a=set(a)
    a=list(a)
    a=sorted(a)
    d=0 
    for i in range(len(a)-1):
        b= a[i+1]-a[i]
        if d>b/2:
            d=d
        else:
            d=math.ceil(b/2)
        

    
    answer = d
    
    return answer
l=15
v=[15,5,3,7,9,14,0]
# l=5
# v=[2,5]

solution(l,v)
```
